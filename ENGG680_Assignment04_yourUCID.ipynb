```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a 3x3 grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

# List of column names (adjust these based on your dataset)
columns = df.columns[:9]

# Loop through the columns and plot histograms with KDE
for i, ax in enumerate(axes.ravel()):
    sns.histplot(df[columns[i]], kde=True, ax=ax)
    ax.set_title(columns[i])
    ax.set_xlabel('Values')
    ax.set_ylabel('Frequency')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()



```

Loaded dataset from the CSV file using pd.read_csv().

```python
# Create a 3x3 grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

# List of column names (adjust these based on your dataset)
columns = df.columns[:9]

# Define colors for histogram and KDE plot
hist_color = "lightgreen"
kde_color = "red"

# Loop through the columns and plot histograms with KDE
for i, ax in enumerate(axes.ravel()):
    sns.histplot(df[columns[i]], kde=True, ax=ax, bins=20, color=hist_color)
    sns.kdeplot(df[columns[i]], ax=ax, color=kde_color)
    ax.set_title(columns[i])
    ax.set_xlabel('Values')
    ax.set_ylabel('Density')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## **2.1 Operating Revenue and Expenses (15 pts)**

```python
# Your Contribution
# Filter the DataFrame to include rows with "Operating revenue" or "Operating expenses" in the "Summary statistics" column
revenue_expenses = df[df['Summary statistics'].isin(["Operating revenue", "Operating expenses"])]

# Filter "GEO" to include all provinces except "Canada"
revenue_expenses = revenue_expenses[revenue_expenses['GEO'] != 'Canada']

# Sort the DataFrame by "GEO"
revenue_expenses = revenue_expenses.sort_values(by="GEO")

# Round all numbers to three decimal places
revenue_expenses = revenue_expenses.round(3)

# Display the resulting DataFrame
print(revenue_expenses)
```

<font color='Green'><b>Loaded  DataFrame from a CSV file

I filtered the DataFrame revenue_expenses to include rows where the "Summary statistics" column contains either "Operating revenue" or "Operating expenses."

I further filtered the "GEO" column to exclude rows with "Canada."

Sorted the DataFrame by the "GEO" column in ascending order.

Rounded up all numerical values in the DataFrame to three decimal places using .round(3).

Finally, printed the resulting DataFrame that meets the specified criteria..</b></font>

Create a bar plot using sns.barplot() with "GEO" on the x-axis, "VALUE" on the y-axis, and "Summary statistics" for hue.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of the plot
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 6))  # Set the figure size
sns.barplot(x="GEO", y="VALUE", hue="Summary statistics", data=revenue_expenses)

# Customize the plot
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title("Operating Revenue and Expenses by Province")
plt.xlabel("Province")
plt.ylabel("Value")
plt.legend(title="Summary Statistics")

# Show the plot
plt.tight_layout()
plt.show()
```
